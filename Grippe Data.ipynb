# https://www.data.gouv.fr/fr/datasets/lieux-de-vaccination-contre-la-grippe-pharmacies-sante-fr/

```python
import pandas as pd

df = pd.read_csv(r'https://www.data.gouv.fr/fr/datasets/r/479afe3e-a2aa-4519-b509-ba9c106a5576', sep=';')
df
```

```python
df[['Adresse_latitude', 'Adresse_longitude']].isna().sum()
```

```python
df = df.dropna(subset=['Adresse_latitude','Adresse_longitude'])
```

```python
df
```

```python
import geopy
from geopy.geocoders import BANFrance
```

```python
locator = BANFrance(user_agent="test")

coordinates = "48.837516, 2.391121"
location = locator.reverse(coordinates)
location.raw['properties']['context'].split(',')[2]
```

```python
locator = BANFrance(user_agent="test")

coordinates = "15.881647, -61.314753"
location = locator.reverse(coordinates)
location.raw['properties']['context'].split(',')[1]
```

```python
len(location.raw['properties']['context'].split(','))
```

```python
def get_region(lat, lon):
    locator = BANFrance(user_agent="test")
    coordinates = str(lat) + ',' + str(lon)
    location = locator.reverse(coordinates)
    
    if len(location.raw['properties']['context'].split(',')) == 3:
        return location.raw['properties']['context'].split(',')[2]
    
    if len(location.raw['properties']['context'].split(',')) == 2:
        return location.raw['properties']['context'].split(',')[1]
    
    else:
        return
```

```python
df.reset_index(inplace=True)
```

```python
df['region'] = " "
```

```python
for k in range(len(df)):
    if df['region'][k] == " ":
        df['region'][k] = get_region(df['Adresse_latitude'][k], df['Adresse_longitude'][k])
    else:
        pass
```

```python
df['region']
```

```python
df.to_csv('App/flu_vaccination.csv', encoding='utf-8')
```

# https://www.sentiweb.fr/france/fr/?page=table

```python
import pandas as pd

df2 = pd.read_csv(r'https://www.sentiweb.fr/datasets/incidence-PAY-3.csv', sep =',',skiprows = 1, encoding = 'iso-8859-1')
df2
```

```python
df2 = df2.sort_values(by = ['week'])
df2.reset_index(inplace=True)
df2
```

```python
str(df2['week'][0])[4:6]
```

```python
def get_year(week): 
    return str(week)[:4]

df2['year'] = df2['week'].map(get_year)
```

```python
def get_nweek(week): 
    return str(week)[4:6]

df2['nweek'] = df2['week'].map(get_nweek)
```

```python
df2['week2'] = df2['week'].map(get_year) + ' - week '+ df2['week'].map(get_nweek)
```
